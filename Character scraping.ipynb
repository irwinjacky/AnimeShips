```python
from bs4 import BeautifulSoup
import time
import urllib.request
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_My_Hero_Academia_characters'

with urllib.request.urlopen(url) as response:
    html = response.read()
    html = html.decode('utf-8')
    
with open ('character.html', 'w') as new_file:
    new_file.write(html)
```

```python
character_soup = BeautifulSoup(html, 'html.parser')
character_soup.prettify
```

```python
mha_characters = character_soup.find_all(class_ = 'vanchor-text')
```

```python
mha_characters = [tag.get_text() for tag in mha_characters]
```

```python
names = pd.DataFrame(mha_characters)
```

```python
names.to_csv('mha_names')
```
