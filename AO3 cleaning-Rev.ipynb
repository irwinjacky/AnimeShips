```python
import pandas as pd

mha_characters = pd.read_csv('mha_names')

characters = pd.read_csv('charactersviadate.csv')
characters
```

```python
character_mha = []
def pullcharacter(names, char_date):
    character_mha.append(names)
    names = pd.DataFrame([s for s in char_date if any(xs in s[0] for xs in names)])
    names.columns = ['name', 'date']
    print(len(names['name']))
    return(names)
```

```python
char_date = characters.values.tolist()
char_date
```
