```python
ships_mha = []
def pullcharacter(names, ship_date):
    ships_mha.append(names)
    names = pd.DataFrame([s for s in ship_date if any(xs in s[0] for xs in names)], columns = ['ships', 'Date'])
    print(names['ships'].value_counts())
    return(names)
```
